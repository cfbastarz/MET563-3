# Equação de Análise Empírica Multivariada

Esta é uma primeira aproximação para o problema de análise multivariado. Isto significa que, na prática, estamos tratando de um conjunto (ou uma série) de observações de quantidades diferentes (eg., temperatura do ar, vento e pressão).

No problema multivariado, procuramos uma solução (análise) que represente o melhor ajustamento do background às observações. No exemplo da Equação de Análise Empírica simples, tínhamos apenas uma equação para uma única variável. Para que possamos extrapolar o problema para $n$ variáveis, será necessário calcular $n$ análises, uma para cada variável de estado:

\begin{align}
\mathbf{x_{a}^{\textit{n}}} = \alpha^{n}\mathbf{y_{o}^{\textit{n}}} + (1 - \alpha^{n})\mathbf{x_{b}^{\textit{n}}}
\end{align}

Onde,

* $\mathbf{x_{a}^{\textit{n}}}$: é o vetor análise da enésima variável
* $\mathbf{x_{b}^{\textit{n}}}$: é o vetor background da enésima variável
* $\mathbf{y_{o}^{\textit{n}}}$: é o vetor observação da enésima variável
* $\alpha^{n}$: é um vetor com os pesos dado à observação 
* $1 - \alpha^{n}$: é um vetor com os pesos dado ao background)

Olhando para nossa equação de análise empírica, percebemos que os elementos $\mathbf{x_{b}^\textit{n}}$ e $\mathbf{y_{o}^\textit{n}}$ já estão definidos. Mas ainda precisamos determinar o parâmetro $\alpha^{n}$, que é o peso a ser atrubuído às observações. ($1 - \alpha$) é um outro peso que será atribuído aos background. Uma vez determinado o valor de $\alpha^{n}$, determinaremos o fator ($1 - \alpha^{n}$) e consequentemente, o valor de $x_{a}$, ou seja, a análise ou vetor análise, representando da mesma forma que $\mathbf{x_{b}^\textit{n}}$ e $\mathbf{y_{o}^\textit{n}}$:

\begin{align}
\mathbf{x_{a}^{\textit{n}}} = \alpha^{n}\mathbf{y_{o}^{\textit{n}}} + (1 - \alpha^{n})\mathbf{x_{b}^{\textit{n}}}
\end{align}

No caso multidimensional, teremos que calcular o fator $\alpha$ para cada uma das variáveis!

## Determinação de $\alpha^{n}$:

Antes de determinarmos o parâmetro $\alpha$, precisamos saber o que ele é e como pode ser definido. $\alpha$ é um parâmetro que relaciona as medidas das variâncias do background e das observações:

\begin{align}
\alpha^{n} = \frac{\sigma^2_{b,n}}{\sigma^2_{b,n} + \sigma^2_{o,n}}
\end{align}

onde,

* $\sigma^2_{b,n}$: é o enésimo vetor com as variâncias do background
* $\sigma^2_{o,n}$: é o enésimo vetor com as variâncias das observações

Os valores destes vetores são "extraídos" a partir das matrizes de covariância dos erros de background e observação. Então, para calcular $\alpha^{n}$, precisamos determinar as matrizes de covariâncias dos erros de $\mathbf{x_{b}^\textit{n}}$ e $\mathbf{y_{o}^\textit{n}}$.

### Erros $E(\mathbf{x_{b}^\textit{n}})$ e $E(\mathbf{y_{o}^\textit{n}})$

A variância é uma medida de dispersão de uma distribuição. Ela pode ser calculada com base no erro da distribuição. Vamos considerar que os erros dos elementos dos vetores $\mathbf{x_{b}^\textit{n}}$ e $\mathbf{y_{o}^\textit{n}}$ são randômicos, ou seja, não há relação entre os erros dos elementos do vetor observação e entre os elementos do vetor background e também não relação entre os elementos dos dois vetores. Mas iremos manter a distribuição destes erros de forma conhecida, vamos manter uma distribuição próxima à normal de forma que possamos controlar a média e o desvio padrão. Para isso, fazemos:

onde,

* $mdo[t,v,p]$: é a média dos erros da observação ($\mu_{o}^{n}$)
* $mdb[t,v,p]$: é a média dos erros do background ($\mu_{b}^{n}$)
* $dpo[t,v,p]$: é o desvio padrão dos erros da observação ($\sigma_{o}^{n}$)
* $dpb[t,v,p]$: é o desvio padrão dos erros do background ($\sigma_{b}^{n}$)

Como temos variáveis distintas, podemos considerar valores de $mdb$ e $dpb$ diferentes entre si. Consequentemente, calcularemos $erro$ e $errb$ para cada uma das variáveis!

Podemos testar alguns valores para os parâmetros acima e ver o que acontece com o ajustamente final da análise de cada variável considerada:

```python
# Média dos erros e desvio padrão da Temperatura
mdot = 1;   dpot = 1.5
mdbt = 2.5; dpbt = 4

# Média dos erros e desvio padrão do Vento
mdov = 1;   dpov = 1.5
mdbv = 3.2; dpbv = 1.5

# Média dos erros e desvio padrão da Pressão
mdop = 1;   dpop = 1.5
mdbp = 2.8; dpbp = 2.1
```

```python
plt.figure(figsize=(14, 4))

plt.subplot(131)
errot = mdot + dpot * np.random.randn(len(x0))
errbt = mdbt + dpbt * np.random.randn(len(x0))
plt.plot(x0, errbt, color="blue", label="$errb_{t}$")
plt.scatter(x0, errot, color="red", label="$erro_{t}$")
plt.title("Erro da Temperatura\n$mdo=1$; $dpo=1,5$\n$mdb=2,5$; $dpb=4$", fontweight="bold", size=15)
plt.legend()
plt.grid(True, which="both")
plt.axhline(y=0, color="k")

plt.subplot(132)
errov = mdov + dpov * np.random.randn(len(x0))
errbv = mdbv + dpbv * np.random.randn(len(x0))
plt.plot(x0, errbv, color="blue", label="$errb_{v}$")
plt.scatter(x0, errov, color="red", label="$erro_{v}$")
plt.title("Erro do Vento\n$mdo=1$; $dpo=1,5$\n$mdb=3,2$; $dpb=1,5$", fontweight="bold", size=15)
plt.legend()
plt.grid(True, which="both")
plt.axhline(y=0, color="k")

plt.subplot(133)
errop = mdop + dpop * np.random.randn(len(x0))
errbp = mdbp + dpbp * np.random.randn(len(x0))
plt.plot(x0, errbp, color="blue", label="$errb_{p}$")
plt.scatter(x0, errop, color="red", label="$erro_{p}$")
plt.title("Erro da Pressão\n$mdo=1$; $dpo=1,5$\n$mdb=2,8$; $dpb=2,1$", fontweight="bold", size=15)
plt.legend()
plt.grid(True, which="both")
plt.axhline(y=0, color="k")

plt.tight_layout()
```

Nosso objetivo é calcular os pesos (fator $\alpha^{n}$) para cada uma das variáveis. $\alpha^{n}$ depende das variâncias dos erros de observação e background. As variâncias dependem dos erros, e já calculamos estes erros.

Portanto, vamos calcular as variâncias de $\mathbf{x_{b}^\textit{n}}$ e $\mathbf{y_{o}^\textit{n}}$. Sabemos que as variâncias de $\mathbf{x_{b}^\textit{n}}$ e $\mathbf{y_{o}^\textit{n}}$ são extraídas das matrizes de covariâncias dos erros de $\mathbf{x_{b}^\textit{n}}$ e $\mathbf{y_{o}^\textit{n}}$ - **por que**?

```python
B = np.cov(xberr)
```

```python
B
```

```python
R = np.cov(yerr)
```

```python
xbvar = np.diag(B)
```

```python
yvar = np.diag(R)
```

```python
alphat = xbvar[0] / (xbvar[0] + yvar[0])
```

```python
alphav = xbvar[1] / (xbvar[1] + yvar[1])
```

```python
alphav
```

```python
alphap = xbvar[2] / (xbvar[2] + yvar[2])
```

## Cálculo de $\mathbf{x_{a}^\textit{n}}$

Observando novamente a equação da análise, notamos que todos os parâmetros estão determinados:

\begin{align}
\mathbf{x_{a}^{\textit{n}}} = \alpha^{n}\mathbf{y_{o}^{\textit{n}}} + (1 - \alpha^{n})\mathbf{x_{b}^{\textit{n}}}
\end{align}

Onde,

* $\mathbf{x_{a}^{\textit{n}}}$: é o vetor análise da enésima variável
* $\mathbf{x_{b}^{\textit{n}}}$: é o vetor background da enésima variável
* $\mathbf{y_{o}^{\textit{n}}}$: é o vetor observação da enésima variável
* $\alpha^{n}$: é um vetor com os pesos dado à observação 
* $1 - \alpha^{n}$: é um vetor com os pesos dado ao background)

```python
xat = alphat * yo1 + (1 - alphat) * xb1
```

```python
xat
```

```python
xav = alphav * yo2 + (1 - alphav) * xb2
```

```python
xav
```

```python
xap = alphap * yo3 + (1 - alphap) * xb3
```
